```python
from openai.error import RateLimitError
from scipy.sparse import data
import time
openai.api_key = ""
def simpleMyPrompt(prompt,input):

  response = openai.Completion.create(
  model="text-davinci-003",
  prompt=prompt,
  temperature=0.7,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
  )
  return response

a=0
data_test1=[]
for i in range(1,4):
  if i*50<data_train["text"].size:
      data_test1.append(data_train['text'].iloc[a:i*40].apply(lambda x:simpleMyPrompt('Provide me with interpretetion of this joke',x)))
  a=i*50
  time.sleep(30)
  print(i)
  #print(data_test1)
```

```python
data_test3 = pd.DataFrame()

data_test3['inter_gpt3_my_full']=[smt for sublist in data_test1 for smt in sublist]
print(data_test3['inter_gpt3_my_full'])
data_test3['inter_gpt3_my']=data_test3['inter_gpt3_my_full'].str['choices'].str[0].str['text'].str.strip()
data_test3['inter']=data_test3['inter_gpt3_my'].str.split('\n')
data_test3
```
